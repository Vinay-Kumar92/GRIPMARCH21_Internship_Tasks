## Step1 : Importing all the required packages/libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step2 : Loading the dataset

```python
dataset = pd.read_csv('SampleSuperstore.csv')
dataset.head()
```

 **Conclusion : There are no null values in the column dataset**

```python
dataset.shape
```

 **Conclusion : We have 13 columns and 9994 records in this dataset**

```python
dataset.columns
```

>### Duplicates in the dataset

```python
dataset.duplicated().sum()
```

 **Conclusion : There are 17 duplicate values in the dataset**

```python
plt.figure(figsize = (10,10))
plt.pie(dataset['Category'].value_counts(), labels = dataset['Category'].value_counts().index, autopct = '%.2f%%', 
        explode = [0.1,0,0])
plt.show()
```

   **Conclusion : The superstore sells their products in 3 categories namely Office Supplies, Furniture and Technology**

```python
plt.figure(figsize = (10,10))
plt.pie(dataset['Ship Mode'].value_counts(), labels = dataset['Ship Mode'].value_counts().index, autopct = '%.2f%%', 
        explode = [0.1,0,0,0])
plt.show()
```

  **Conclusion : The superstore delivers the products by 4 modes of shipment. And they are 'First Class', 'Second Class', 'Standard Class' and 'Same day'. The 'Standard Class' ship mode is prefered by the most by the customers**

* The segment "Home" has most to total sales as compared to the other 2 segments i.e coperate and home office
